```python
from __future__ import division
import sklearn as skl
import pandas as pd
import numpy as np
from matplotlib import pylab as pl
from matplotlib import pyplot as plt
%matplotlib inline
#pl.rcParams['figure.figsize']= 16,5
```

Leyendo Datos

```python
data = pd.read_csv('resources/Consulta_Banxico.csv',encoding='latin1', sep=',', skiprows=17, skip_blank_lines=True)
data.columns = ['Fecha','Valores']
#data.set_index('Fecha',inplace=True)
```

Funciones de Normalización

```python
def normalize_func(x,mean,std):
    norm = (x-mean)/std
    return norm

def normalize_data(data):
    mean = data['Valores'].mean()
    std = data['Valores'].std()
    temp=[]
    for x in data['Valores']:
        temp.append(normalize_func(x,mean,std))
    norm_data = data.copy()
    norm_data['Valores'] = temp
    norm_data.columns = ['Fecha','Valores_norm']
    del temp
    return norm_data

def ndarray_toArray(values):
    temp = [x[0] for x in values]
    return temp
```

Funciones de Visualización

```python
def plot_timeSerie(values,ticks):
    fig =plt.figure()
    ax=fig.add_subplot(111)
    ax.plot(values,'.-')
    plt.xticks(range(0,len(values),12),rotation=90)
    plt.show()
```

```python
norm_data = normalize_data(data)
```

```python
plot_timeSerie(norm_data['Valores_norm'],norm_data['Fecha'].apply(str))
```
